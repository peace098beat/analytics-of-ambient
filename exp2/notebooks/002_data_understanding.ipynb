# Plot waveform

```python
def plot_wave(wave_path):
    print(wave_path)
    woodpecker = Audio(wave_path, )
    woodpecker.signal.plot_2D_Data(
        "time",
        type_plot="curve",
        color_list=COLORS,
    )

    return woodpecker
```

```python
%%time

for w in wav1_paths:
    plot_wave(w)
```

```python
%%time

for w in wav2_paths:
    plot_wave(w)
```
